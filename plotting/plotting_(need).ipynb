## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed',linewidth = 9,label = 'Y values')  #This is the "curve" itself
                                                   #(<xvalues>,<yvalues>,<colour>,<linestyle>,<linewidth>,<label>)
plt.xlabel('x-values')  #x-axis name
plt.ylabel('y-values')  #y-axis name
plt.title('X vs Y')     #title name
plt.grid(alpha=0.23)    #alpha = shade of grid line
plt.legend(loc='upper left') #specify location of legend
plt.show()  #show function
```

### 1.2 Adding another plot

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x,y_1,color='red',linestyle='dashed',linewidth=1,label='y_1 values')
plt.plot(x,y_2,color='blue',linestyle = 'solid',linewidth=1,label='y_2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25)
plt.legend(loc = 'upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = 0.2, 1

plt.plot(x,y_1,c = 'red',ls= 'dashed',label = 'y_1 values')
plt.plot(x,y_2,c='blue',ls='solid',label = 'y_2 values')
plt.errorbar(x,y_3,xerr = x_error, yerr = y_error,c = 'black',label = 'y_3 values with errors')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("X vs Y")
plt.grid(alpha = 0.25)
plt.legend(loc = 'upper left')
plt.show()

#Note that you can also provide a list of erros so each point has its own


```

### 1.5 Better with NumPy

```python
# Often it is easier to use NumPy arrays instead of Python lists
import numpy as np

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(x,np_y_1,c='black',ls = 'dashed',label = 'love')
plt.plot(x,np_y_2,c = 'green', ls = 'solid', label = 'hate')
plt.xlabel('bloop')
plt.ylabel('whloop')
plt.title('hehe')
plt.grid(alpha = 0.5,c = 'red')
plt.legend(loc = 'best')
plt.show()
```

### 1.6 Adding mathematical functions

```python
# One of the advantages of NumPy arrays is that they can easily generate data-related mathematical functions.

x = np.array([i/10 for i in range(15)])
y = x**2
z = np.sin(x)

plt.plot(x,y,c='red',ls = 'solid',label = 'x^2')
plt.plot(x,z,c = 'blue',ls = 'solid',label = 'cos x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('X vs Y')
plt.grid(alpha= 0.5, c = 'black')
plt.legend(loc = None)
plt.show()

```

### 1.7 We need another axis!

```python
#plt.twinx()  allows us to have another y-axis that shares the same x-axis

x = np.array([i/10 for i in range(15)])
y = x**2
z = np.sin(x)



plt.plot(x,y,c='red',ls = 'solid',label = 'x^2')
plt.legend(loc = 'lower right')
plt.ylabel('y-axis')

plt.twinx()
plt.plot(x,z,c = 'blue',ls = 'solid',label = 'cos x')



plt.ylabel('y-axis')
plt.title('X vs Y')
plt.grid(alpha= 0.5, c = 'black')
plt.legend(loc = 'upper left')
plt.show()

#np.linspace(x,y,z) -- this function is useful when you have less data points
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
# The first thing you should do with any data is to look at it with a simple plot. (Advice)

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date,anomaly, alpha = .5)
plt.ylim([-8,8])

# The hlines() function needs a y-value and starting and ending values for x

# plt.hlines(<y-value>, <start-x>, <end-x>, <linestyle>, <colours>)

plt.hlines(0, date[0], date[-1], ls= 'dashed', color = 'red')
```

```python
#For plotstyle --> http://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html

plt.style.use('dark_background')

plt.plot(date, anomaly, alpha = .5)

plt.hlines(0,date[0],date[-1],ls = 'dashed',color = 'blue')

plt.show()
```

```python
# If you want to reset things jump out of this style, you need to set the default style using:
plt.style.use('default')
```

```python
plt.style.use('dark_background')

plt.plot(date,anomaly,ls = 'solid', lw = 0.3, color = 'red')
plt.hlines(0,date[0],date[-1],color = 'white',alpha = 1)
plt.ylim([-8,8])
plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')

plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
from numpy import *

x = np.array([i for i in range(51)])
y = np.array(e**(-x/10)*sin(x))


plt.plot(x,y, color = 'blue')
plt.ylim([-1,1])
plt.xlabel('x', fontsize = 15, color = 'darkblue')
plt.ylabel('y', fontsize = 15, color = 'darkblue')
plt.grid(alpha =.25)
plt.title(r'$e^{-x/10}sin(x)$', fontsize = 20)


plt.show()




```
