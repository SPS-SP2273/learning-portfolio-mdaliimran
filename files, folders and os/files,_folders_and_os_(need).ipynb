### 1.4 Opening and closing files

```python
with open('spectrum-01.txt', 'r') as file:
    file_content = file.read()

print(file_content)

#This is what you would typically do to read a file

#open function opens a stream in your file

#r specifies read only

#by using with, we dont have to bother closing the file




```

```python
text = 'Far out in the uncharted backwaters of the unfashionable end of the western spiral arm of the Galaxy lies a small unregarded yellow sun.\nOrbiting this at a distance of roughly ninety-two million miles is an utterly insignificant little blue green planet whose ape-descended life forms are so amazingly primitive that they still think digital watches are a pretty neat idea.'

with open('my-text-once.txt', 'w') as file:
    file.write(text)
    
#w specifies write

#note that if the file does not exist, this will create the file

#if it does exist, it rewrites

with open('my-text-once.txt','r') as file:
    x = file.read()
print(x)    

#-----------------------------------

with open('my-text-lines.txt', 'w') as file:
    for line in text.splitlines():
        file.writelines(line)

#This method is much slower but can be useful
```

### 2.4 Checking for existence

```python
#Soemtimes, when you're trying to create a new folder, it will run an error stating the folder already exist


try:
    os.mkdir('fruits')
    for fruitz in ['apple','banana','orange']:
        path = os.path.join('fruits',fruitz)
        os.mkdir(path)
except:
    print(f'{path} is already existent.')

```

```python
#Another method to check

for fruitz in ['apple','banana','orange']:
    path = os.path.join('fruits',fruitz)

if os.path.exists(path):
    print(f'{path} already exist')
    
else:
    os.mkdir(path)
```

## Exercise 8 :  Tidying a collaboration

```python
helper = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

change = helper['oct']

x = 'oct-08_singapore.txt'

x = x.replace('oct',change)

print(x)


```

```python
def rename_my_file(old_file_name):
    helper = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'}
    
    month = old_file_name[:3]
    change = helper[month]
    new_file_name = old_file_name.replace(month,change)
    
    return new_file_name
    
    
```

```python
small_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

result = [rename_my_file(i) for i in small_list]

print(result)
    
```

```python
for i in glob.glob('*.txt'):
    os.rename(i, rename_my_file(i))
    
glob.glob('*.txt')    
```

```python
liz = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']


for countries in liz:
    path = os.path.join('os-collaboration-exercise-data',countries)
    os.mkdir(path)
    
```

```python
list_cairo = glob.glob('*cairo*')
for item in list_cairo:
    path_to_destination = os.path.join('Cairo',item)
    shutil.copy(item,path_to_destination)
    
```

```python
#Verifying the previous step

print(glob.glob('Cairo/*'))
```

```python
new_list = ['Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for countries in new_list:
    list_country = glob.glob(f'*{countries.lower()}*')
    for item in list_country:
        path_to_destination = os.path.join(f'{countries}',item)
        shutil.copy(item, path_to_destination)
```
